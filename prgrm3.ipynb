```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(15,15))
#plot 1:

xpoints= np.array([1,6])
ypoints = np.array([3, 8])
plt.subplot(3,3,1)
plt.xlabel("x-axis",color='blue')
plt.ylabel("y-axis",color='blue')
plt.plot(xpoints, ypoints,'o')
plt.title("TITLE 1")

#plot 2:
xpoints = np.array([3,8,1,10])
ypoints = np.array([6,2,8,1])
plt.subplot(3,3,2)
plt.plot(ypoints, marker = '*',color='orange',linewidth = '2.5')
plt.plot(xpoints, linestyle='dashed',color='blue')


#plot 3:
ypoints = np.array([3, 8, 1, 10])
plt.subplot(3,3,3)
plt.plot(ypoints, marker = 'o',ms=8, mfc='yellow',mec='black')


#plot 4:
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.subplot(3,3,4)
plt.plot(x, y)
plt.grid()



#plot 5:
x = np.array([5,7,8,7,2,17,2,9,4,11])
y = np.array([99,86,87,88,111,86,103,87,94,78])
plt.subplot(3,3,5)
colors = np.array(["red","green","yellow","blue","orange","black","hotpink","violet","brown","purple"])
plt.scatter(x, y, c=colors,alpha=0.5)


#plot 6:
x = np.array([1,5,9])
y = np.array([5,2,8])
plt.subplot(3,3,6)
plt.bar(x,y,color='r',width=1)


#plot 7:
y = np.array([35, 20, 20, 25])
plt.subplot(3,3,7)
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
plt.pie(y, labels = mylabels, startangle = 90 ,explode=myexplode,shadow='true')
plt.legend()
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# Example training data
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 4, 12, 16, 25])
classes = np.array([0, 0, 1, 1, 0])  # Example class labels

# Combine x and y for training data
X_train = np.column_stack((x, y))

# Define and train the kNN classifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, classes)

# New point to predict
new_x = 3
new_y = 10
new_point = np.array([[new_x, new_y]])

# Make prediction
prediction = knn.predict(new_point)

# Plotting
plt.scatter(x, y, c=classes, label='Training Data')
plt.scatter(new_x, new_y, c=prediction, marker='X', s=200, label='New Point')
plt.text(x=new_x-1.0, y=new_y-0.9, s=f"new point, class: {prediction[0]}", fontsize=12)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
